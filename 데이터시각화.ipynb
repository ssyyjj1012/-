## 연습문제
- 다음 공식을 NumPy 배열로 나타내고 그래프로 그려보자
 - 1차원식 : f1(x) = 2x + 3
 - 2차원식 : f2(x) = x^2*PI
 - 3차원식 : f3(x) = x^3


### 1차원식

```python
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
  return 2 * x + 3

arr1 = np.arange(-10,10,0.1)
arr2 = f1(arr1)

plt.plot(arr1,arr2)
plt.show()
```

### 2차원식

```python
def f2(x):
  return x * x * np.pi

arr1 = np.arange(-10,10,0.1)
arr2 = f2(arr1)

plt.plot(arr1,arr2)
plt.show()
```

```python
def f3(x):
  return x * x * x

arr1 = np.arange(-10,10,0.1)
arr2 = f3(arr1)

plt.plot(arr1,arr2)
plt.show()
```

# 03 그래프 꾸미기
- matplotlib에서 plot를 사용하여 그래프를 다양하게 꾸밀 수 있음
- 색상, 마커, 라인스타일, 드로우스타일, 축 이름, 타이틀 등 다양한 꾸미기 가능



- 꾸미기에 사용할 배열 두개 생성

```python
import matplotlib.pyplot as plt
import numpy as np

arr1 = np.random.randn(10)
plt.plot(arr1)
arr2 = np.random.randn(10)
plt.plot(arr2)
plt.show()
```

- b : blue

- c : cyan
- m : magenta

- k : black

```python
plt.plot(arr1, color = 'b')
plt.plot(arr2, color = 'r')
plt.show()
```

### 그래프 꾸미기 - 마커
- 마커는 그래프 상에 찍히는 좌표를 말함
- 일반적으로는 잘 보이지 않으나, 그래프가 꺾일 때는 명확히 인식 가능
- 마커를 꾸밈으로써 데이터의 정확하 위치 파악 가능

```python
arr1 = np.random.randn(10)
arr2 = np.random.randn(10)

plt.plot(arr1, color = 'b', marker = 'o')
plt.plot(arr2, color = 'r', marker = '*')
plt.show()
```

### 그래프 꾸미기 - 라인스타일
- 라인스타일은 서늬 형태를 정함
- 선, 점선 등 여러선의 형태 구현


```python
plt.plot(arr1, color = 'b', marker = 'o', linestyle = '--')
plt.plot(arr2, color = 'r', marker = '*', linestyle = '-.')
plt.show()
```

### 그래프 꾸미기 - 드로우스타일
- 드로우스타일은 꺾은 그래프 그릴 때 사용
- 드로우 스타일은 다음과 같은 종류를 가지며, 이 종류에 따라 꺾는 부분 정함 
 - default
 - steps
 - steps-pre
 - steps-mid
 - steps-post

```python
plt.plot(arr1, color = 'b', drawstyle = 'default')
plt.plot(arr1, color = 'r', drawstyle = 'steps')
plt.show()
plt.plot(arr1, color = 'g', drawstyle = 'steps-pre')
plt.plot(arr1, color = 'y', drawstyle = 'steps-mid')
plt.plot(arr1, color = 'k', drawstyle = 'steps-post')
plt.show()
```

### 그래프 꾸미기 - 단축코드
- 색상 마커 라인스타일 순으로 문자열을 결합하여 두번째 인자로 넘겨줄 수 있음

```python
plt.plot(arr1, 'bo--', drawstyle = 'steps-post')
plt.plot(arr2, 'r*-.', drawstyle = 'steps')
plt.show()
```

# 04 서브플롯 - 피규어(figure)
- figure는 도표를 담는 객체
- plt.figure() 함수를 사용하여 객체 생성

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

fig = plt.figure()
fig
```

### 피규어 나누기
- fig의 객체는 add_subplot 이라는 함수를 사용하여 여러 개의 도표를 그림
- subplot은 하나의 도표를 여러 조각으로 나눠 사용
- 조각을 나누는 기준을 첫 번째 인자와 두 번째 인자로 전달
- 세 번째 인자에는 나눈 도표의 몇 번째 부분(조각)을 사용할 지 전달
- 서브플롯을 생성하여 plt.show()를 사용하여 출력
- add_subplot(2,2,1) -> 2 x 2 로 figure을 나누고 그 중 첫번째 부분을 사용

```python
fig1 = plt.figure()
fig1.add_subplot(2,2,1) # 2x2 크기의 표에 1번 위치에 빈 표 생성
plt.show()
```

### 피규어 하나로 사용하기 
- 플롯을 하나로 사용하고 싶은 경우에는 1,1,1을 입력

```python
fig1 = plt.figure()
fig1.add_subplot(1,1,1) # 1x1 크기의 표에 1번 위치에 빈 표 생성
plt.show()
```

### 피규어에 그래프 그리기
- 피규어에 그래프를 그릴 때는 두가지 방법이 있습니다.

 1) 기존에 사용하던 plt.plot 함수 실행

 2) subplot 객체를 받아 plot 함수 실행

```python
## plt.plot 함수로 그리기
fig2 = plt.figure()
ax1 = fig2.add_subplot(2,1,1)
plt.plot(np.random.randn(50))

## subplot.plt 함수로 그리기
ax2 = fig2.add_subplot(2,1,2)
ax2.plot(np.random.randn(100))
plt.show()
```

- subplots 함수를 사용하면 리스트 형태로 서브 플롯을 받아서 사용
- 첫 번째 인자와 두 번째 인자는 이전과 같은 용도로 사용
- for 문을 사용하여 출력하면 다수의 그래프를 빠르게 그림

```python
fig, axes = plt.subplots(2,2)

for i in range(2):
  for ll in range(2):
    axes[i,ll].plot(np.random.randn(50*(i+1)*(ll+1)).cumsum(),'k-')

plt.show()
```

### 간격 조절
- 서브플롯 간의 간격을 조절하기 위해 다음 함수 사용
 - plt.subplots_adjust(wspace = 0 , hspace = 0)
- wspace에는 너비 간격, hspace에는 높이 간격을 각각 전달


```python
fig, axes = plt.subplots(2,2)

for i in range(2):
  for ii in range(2):
    axes[i,ii].plot(np.random.randn(50*(i+1)*(ii+1)).cumsum(), 'k-')

plt.subplots_adjust(wspace = 0, hspace = 0)
plt.show()
```


- x축과 y축을 공유하는 기능 사용
- subplots 함수에 sharex, sharey 를 전달하여 활성화

```python
fig, axes = plt.subplots(2,2, sharex = True, sharey = True)

for i in range(2):
  for ii in range(2):
    axes[i,ii].plot(np.random.randn(50*(i+1)*(ii+1)).cumsum(), 'k-')

plt.subplots_adjust(wspace = 0, hspace = 0)
plt.show()
```

### 연습문제
다음 공식을 NumPy 배열로 나타내고 하나의 도표에 나눠서 그래프를 그려보자

- 1차원식 : f1(x) = x      위치 (0,0)
- 2차원식 : f2(x) = x**2   위치 (0,1)
- 3차원식 : f3(x) = x**3   위치 (1,0)
- 4차원식 : f4(x) = x**4   위치 (1,1)

```python
import numpy as np
import matplotlib.pyplot as plot

arr1 = np.arange(-10, 10, 0.1)

fig, axes = plt.subplots(2,2, sharex = True)
```

```python
def f1(x,n):
  return x**n

n = 0

for i in range(2):
  for ii in range(2):
    n+=1
    axes[i,ii].plot(arr1, f1(arr1,n))

plt.subplots_adjust(hspace = 0, wspace = 0.5)
plt.show()
```
